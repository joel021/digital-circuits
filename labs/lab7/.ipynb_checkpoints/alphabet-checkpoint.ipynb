### Alphabet Definition

```python
"""
let {sR:fR}{sG:fG}{sB:fB} = RGB_i coding to colors of one pixel i. fR-sR = fG-sG = fB-sB = 4 bits.
let one line of the file defined as {R,G,B}.
One character is constructed with 16x16 pixels.
256 positions on memory is used to store a character.
"""
class Alfhabet ():
    
    def __init__(self):

        self.map = {
            " ": self.space,
            "A": self.A,
            "B": self.B,
            "C": self.C,
            "D": self.D,
            "E": self.E,
            "F": self.F,
            "G": self.G,
            "H": self.H,
            "I": self.I,
            "J": self.J,
            "K": self.K,
            "L": self.L,
            "M": self.M,
            "N": self.N,
            "O": self.O,
            "P": self.P,
            "Q": self.Q,
            "R": self.R,
            "S": self.S,
            "T": self.T,
            "U": self.U,
            "V": self.V,
            "W": self.W,
            "X": self.X,
            "Y": self.Y,
            "?": self.heart
        }

        self.char_encoding = dict()

        l = math.ceil(math.log(len(self.map.keys()),2))

        for i, k in enumerate(self.map.keys()):
            self.char_encoding[k] = bin(i)[2:]
            comp_arr = np.zeros(l-len(self.char_encoding[k])).astype(int)
            self.char_encoding[k] =  str(comp_arr)[1:-1].replace(" ", "") + self.char_encoding[k] + "\n"
            
    def space(self, x, y):
        return "000000000000\n"

    def A(self, x, y):
        r1 = [math.ceil(-2*x+14), math.ceil(15-2*x), math.ceil(16-2*x)]
        r2 = [math.ceil(2*x-14), math.ceil(2*x-15), math.ceil(2*x-16)]
        r3 = [8,9,10]

        if y in r1 or y in r2 or (y in r3 and y >= r1[1] and y >= r2[1]):
            return "111111111111\n" #"111111111111\n" #max color in RGB with 4 bits each.
        else:
            return "000000000000\n" #"000000000000\n"

    def B(self, x, y):
        y_t = [13,14] #top of B
        y_b = [1,2] #y base of B
        x_b = [1,2] #x base of B
        y_m = [6,7]
        x_tt = np.ceil([(10 + math.cos(y/2+1)), 11 + math.cos(y/2+1), 12 + math.cos(y/2+1)])
        x_tb = np.ceil([10 + math.cos(y / 2 + 4), 11 + math.cos(y / 2 + 4), 12 + math.cos(y / 2 + 4)])

        if (y in y_m and x >= 1 and x <= 10) or (y in y_t and x >= 1 and x <= 10) or (y in y_b and x >= 1 and x <= 10) or (x in x_b and y >= 1 and y <= 14)  or (x in x_tt and y <= 13 and y >= 7) or (x in x_tb and y <= 7 and y >= 1) :
            return "111111111111\n"#"111111111111\n"
        else:
            return "000000000000\n"#"000000000000\n"
    
    def C(self, x, y):
        c1 = (x - 7.5)**2 + (y / 1.2 - 6)**2 <= math.ceil(5.5**2)
        c2 = (x - 7.5)**2 + (y / 1.2 - 6)**2 >= math.ceil(4.5**2)

        c3 = y >= 11 or y <= 3 

        if (c1 and c2 and x <= 8) or (c1 and c2 and x >= 8 and c3):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def D(self, x, y):
        c1 = (x - 7.5)**2 + (y / 1.2 - 6)**2 <= math.ceil(5.5**2)
        c2 = (x - 7.5)**2 + (y / 1.2 - 6)**2 >= math.ceil(4.5**2)

        c3 = x >= 1 and x <= 2 and y >= 1 and y <= 14 #vertical base of the D
        c4 = x >= 1 and x <= 8 and y >= 1 and y <= 2 #horizontal base of the D
        c5 = x >= 1 and x <= 8 and y >= 13 and y <= 14

        if (c1 and c2 and x >= 8) or c3 or c4 or c5:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def E(self, x, y):
        c1 = x >= 1 and x <= 14
        c2 = y >= 13 and y <= 14
        c3 = y >= 7 and y <= 8
        c4 = y >= 1 and y <= 2
        c5 = x >= 1 and x <= 2
        c6 = y >= 1 and y <= 14

        if c1 and c4 or c1 and c3 or c1 and c2 or c5 and c6:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def F(self, x, y):
        c1 = x >= 1 and x <= 14
        c3 = y >= 7 and y <= 8
        c4 = y >= 1 and y <= 2
        c5 = x >= 1 and x <= 2
        c6 = y >= 1 and y <= 14

        if c1 and c4 or c1 and c3 or c5 and c6:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def G(self, x, y):
        c1 = (x - 7.5)**2 + (y / 1.2 - 6)**2 <= math.ceil(5.5**2)
        c2 = (x - 7.5)**2 + (y / 1.2 - 6)**2 >= math.ceil(4.5**2)

        c3 = y >= 9 or y <= 5

        c4 = y >= 9 and y <= 10 and x >= 8 and x <= 13 

        if (c1 and c2 and x <= 8) or (c1 and c2 and x >= 8 and c3) or c4:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def I(self, x, y):

        t1 = x >= 6 and x <= 11 and y >= 13 and y <= 14
        t2 = x >= 6 and x <= 11 and y >= 1 and y <= 2

        if x >= 7 and x <= 9 and y >= 1 and y <= 14 or t1 or t2:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def H(self, x, y):

        c1 = x >= 1 and x <= 14 #vertical limitation
        c2 = y >= 1 and y <= 14 #horizontal limitation
        c3 = x >= 1 and x <= 2 #horizontal line 1
        c4 = x >= 13 and x <= 14
        c5 = y >= 6 and y <= 7

        if c1 and c3 or c1 and c4 or c2 and c5:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def J(self, x, y):

        c1 = (x-11)**2+(y-11)**2 <= 25
        c2 = (x-11)**2+(y-11)**2 >= 10
        u = c1 and c2 and y >= 11 and y <= 14

        I = x >= 13 and x <= 14 and y >= 1 and y <= 14

        T = x >= 11 and x <= 14 and y >= 1 and y <= 2

        if u or I or T:
            return "111111111111\n"
        else:
            return "000000000000\n"
        
    def K(self, x, y):

        l = x >= 1 and x <= 2 and y >= 1 and y <= 14
        
        r_s = 9*x+13*y >= 115
        r_i = 9*x+13*y <= 139
        l2 = x >= 1 and x <= 14 and y >= 1 and y <= 14

        r_s1 = 7.5*x-6*y >= -10
        r_s2 = 7.5*x-6*y <= 6
        l3 = x >= 6 and x <= 12 and y >= 7 and y <= 14

        if l or (r_s and r_i and l2) or (r_s1 and r_s2 and l3):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def L(self, x, y):

        l1 = x >= 1 and x <= 12 and y >= 13 and y <= 14

        l2 = x >= 1 and x <= 2 and y >= 1 and y <= 14

        if l1 or l2:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def M(self, x, y):

        u1 = x >= 1 and x <= 2 and y >= 1 and y <= 14

        u2 = (6*x-5*y >= -10 and 6*x-5*y <= 7) and (y >= 1 and y <= 10 and x >= 1 and x <= 8)

        u3 = (x+y >= 14 and x+y <= 17) and (x >= 8 and x <= 14 and y >= 1 and y <= 14)

        u4 = x >= 13 and x <= 14 and y >= 1 and y <= 14

        if u1 or u2 or u3 or u4:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def N(self, x, y):

        u1 = x >= 1 and x <= 2 and y >= 1 and y <= 14

        u2 = (6*x-5*y >= -10 and 6*x-5*y <= 7) and (y >= 1 and y <= 10 and x >= 1 and x <= 8)

        u3 = (x+y >= 14 and x+y <= 17) and (x >= 8 and x <= 14 and y >= 1 and y <= 14)

        if u1 or u2 or u3:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def O(self, x, y):
        c1 = (x - 7.5)**2 + (y / 1.2 - 6)**2 <= math.ceil(5.5**2)
        c2 = (x - 7.5)**2 + (y / 1.2 - 6)**2 >= math.ceil(4.5**2)

        if (c1 and c2):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def P(self, x, y):
        u1 = ((x-7)**2 + (y - 5)**2 <= 16 and (x-7)**2 + (y - 5)**2 >= 4) and x >= 7 and x <= 14 and y >= 1 and y <= 7
    
        u2 = x >= 1 and x <= 8 and y >= 1 and y <= 2

        u3 = x >= 1 and x <= 6 and y >= 7 and y <= 8

        u4 = x >= 1 and x <= 2 and y>= 1 and y <= 14

        if u1 or u2 or u3 or u4:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def Q(self, x, y):
        c1 = (x - 7.5)**2 + (y / 1.2 - 6)**2 <= math.ceil(5.5**2)
        c2 = (x - 7.5)**2 + (y / 1.2 - 6)**2 >= math.ceil(4.5**2)
        u3 = (-6.5*x + 6.5*y <= 20 and -6.5*x + 6.5*y >= 9) and x >= 7 and x <= 14 and y >= 7 and y <= 15

        if (c1 and c2 or u3):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def R(self, x, y):
        u1 = ((x-7)**2 + (y - 5)**2 <= 16 and (x-7)**2 + (y - 5)**2 >= 4) and x >= 7 and x <= 14 and y >= 1 and y <= 7
    
        u2 = x >= 1 and x <= 8 and y >= 1 and y <= 2

        u3 = x >= 1 and x <= 6 and y >= 7 and y <= 8

        u4 = x >= 1 and x <= 2 and y>= 1 and y <= 14

        u5 = -x+y <= 3 and -x+y >= 0 and x >= 7 and x <= 14 and y >= 7 and y <= 14

        if u1 or u2 or u3 or u4 or u5:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def S(self, x, y):
        c1 = (x - 7.5)**2 + (y / 1.2 - 6)**2 <= math.ceil(5.5**2)
        c2 = (x - 7.5)**2 + (y / 1.2 - 6)**2 >= math.ceil(4.5**2)
        u1 = c1 and c2 and ( y >= 1 and y <= 5 or y >= 10 and y <= 14)

        u2 = ( x/3.9- math.tanh(y/1.3-6) >= 1.6 and x / 4.1 - math.tanh(y / 1.3 - 5) <= 2 ) and y >= 5 and y <= 10

        if (u1 or u2):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def T(self, x, y):

        u1 = x >= 1 and x <= 14 and y >= 1 and y <= 2
        u2 = x >= 7 and x <= 8 and y >= 1 and y <= 14

        if u1 or u2:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def U(self, x, y):
        c1 = (x - 7.5)**2 + (y - 7.5)**2 <= math.ceil(6.5**2)
        c2 = (x - 7.5)**2 + (y - 7.5)**2 >= math.ceil(3.5**2)
        u1 = c1 and c2 and y >= 10

        u2 = x >= 1 and x <= 2 and y >= 1 and y <= 10
        u3 = x >= 13 and x <= 14 and y >= 1 and y <= 10

        if (u1 or u2 or u3):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def V(self, x, y):
        u1 = 13*x-6*y <= 30 and 13*x-6*y >= 7 and x >= 1 and x <= 7 and y >= 1 and y <= 14
        u2 = 13*x+7*y >= 161 and 13*x+7*y <= 189 and x >= 7 and x <= 14 and y >= 1 and y <= 14

        if u1 or u2:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def W(self, x, y):

        u1 = 13*x-2.5*y <= 35 and 13*x-2.5*y >= 10.5 and x >= 1 and x <= 4 and y >= 1 and y <= 14
        u2 = -7*x-3.5*y <= -58 and -7*x-3.5*y >= -73 and x >= 3.5 and x <= 7
        u3 = 7*x-3.5*y <= 39 and 7*x-3.5*y >= 24.5 and x >= 7 and x <= 10.5
        u4 = 13*x+3.5*y >= 160 and 13*x+3.5*y <= 185.5 and x >= 10.5 and x <= 14 and y >= 1

        if u1 or u2 or u3 or u4:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def X(self, x, y):
        u1 = 13*x+7*y >= 115 and 13*x+7*y <= 143 and y >= 1 and y <= 14
        u2 = -13*x+7*y >= -64 and -13*x+7*y <= -38.5 and y >= 1 and y <= 14

        if u1 or u2:
            return "111111111111\n"
        else:
            return "000000000000\n"

    def Y(self, x, y):
        u1 = 13*x+7*y >= 115 and 13*x+7*y <= 143 and y >= 1 and y <= 8
        u2 = -13*x+7*y >= -64 and -13*x+7*y <= -38.5 and y >= 1 and y <= 8
        u3 = x >= 7 and x <= 8 and y >= 7 and y <= 14

        if (u1 or u2) or (u3):
            return "111111111111\n"
        else:
            return "000000000000\n"

    def heart(self, x, y):

        u1 = False
        u2 = True
        u3 = True

        if x >= 1 and x <= 14:
            u1 = -y >= -4 * math.sqrt(3 - math.sqrt(abs(2.7*x - 20) / 2)) - 4.4

        if x >= 1 and x <= 7.5:
            u2 = y >= - math.sqrt(-(x - 4.2)**2 + 11) + 4.4
        
        if x >= 7.5 and x <= 14:
            u3 = y >= - math.sqrt(-(x - 10.7)**2 + 11.5) + 4.4

        if u1 and u2 and u3:
            return "111100000000\n"
        else:
            return "000000000000\n"

    def get_RGB(self, char):
        rgb_str = ""
        func = self.map[char]

        for y in range(0,16):
            for x in range(0, 16):
                rgb_str += func(x, y)

        return rgb_str
```

### A letter in RGB encoding

```python
W = 640 #width of the screen
H = 480 #height of the screen
c_w = 16 #width of char
c_h = 16 #heigth of char
qtd_x = 640 / c_w
qtd_y = 480 / c_h

len_smem = 2**math.ceil(math.log(int(qtd_x*qtd_y),2)) #quantity of lines on the screen memory
alphabet_ = Alfhabet()
```

#### Store the characters on BitmapMemory

```python
qtd = len(alphabet_.map.keys())*c_w*c_h
l = 2**math.ceil(math.log(qtd,2))

with open("src/bmem_screen.mem", "w") as f:
    for k in alphabet_.map.keys():
        f.write(alphabet_.get_RGB(k))

    for i in range( int(l/(c_w*c_h) -  len(alphabet_.map.keys()) ) ):
        f.write(alphabet_.get_RGB(" "))
    
```

### Store a text to be displayed on the screen

```python
def find_enter(msg, char_len, W):
    msg_r = ""
    for i in range(len(msg)):
        msg_i = msg[i:i+1]

        if msg_i is "/":
            word_len = len(msg[0:i])#how many chars we already have completed on the screen
            qtd_chars = int(W/char_len)
            remain = word_len % qtd_chars #remaining spaces to complete one line
            char_encoding = ""

            for j in range(qtd_chars-remain):
                char_encoding += " "
                
            msg_r += char_encoding
        else:
            msg_r += msg_i
            
    return msg_r
```

```python
msg = find_enter("  HELLO WORD ? // GCET TWO HUNDRED THIRTY ONE // UFRB // JOEL PIRES", c_w, W)

l = math.ceil(math.log(len(alphabet_.map.keys()),2))
blank = str(np.zeros(l).astype(int))[1:-1].replace(" ", "")

with open("src/smem_screen.mem", "w") as f:
    for char in list(msg):
        f.write(alphabet_.char_encoding[char])

    for i in range(len_smem - len(msg)):
        f.write(blank+"\n")
```
