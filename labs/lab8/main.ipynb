```python
def signed_number(value, length):

    if value >= 0:
        v = bin(abs(value-(1<<length))).replace("0b", "")
    else:
        v = bin(abs(abs(value)-(1<<length))).replace("0b", "")

    return (length - len(v))*"0" + v

def parse_to_bin(instruction):
    #[intr, arg1, arg2, arg3]
    
    if instruction[0] in instr_type["R"]:
        #syntax: [instr, rd, rs, rt]

        rd = bin(int(instruction[0])).replace("0b", "")        
        rd = (5-len(rd))*"0"+rd

        rs = bin(int(instruction[1])).replace("0b", "")        
        rs = (5-len(rs))*"0"+rs

        rt = bin(int(instruction[2])).replace("0b", "")        
        rt = (5-len(rt))*"0"+rt

        return 5*"0" + rs + rt + rd + 5*"0" + instruction_set[instruction[0]]

    elif instruction[0] in instr_type["I"]:
        #syntax: instr rt, rs, imm
        rt = bin(int(instruction[0])).replace("0b", "")        
        rt = (5-len(rt))*"0"+rt

        rs = bin(int(instruction[1])).replace("0b", "")        
        rs = (5-len(rs))*"0"+rs

        imm = signed_number(int(instruction[2]), 16)

        return instruction_set[instruction[0]] + rs + rt + imm
    elif instruction[0] in instr_type["J"]:
        
        pass
```
